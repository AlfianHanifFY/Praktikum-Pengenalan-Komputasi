```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("P05-01.csv")
```

```python
data = df.loc[df["Card_Category"] == "Platinum"]
data[["Customer_Age"]].plot(kind="hist" , bins = [0,20,40,60,80,100] ,rwidth = 0.8)
```

```python
data = df.loc[df["Credit_Limit"] > 5000]
data["Income_Category"].value_counts().plot(kind="pie")
```

```python
data = df.loc[df["Gender"] == "F"]
data.plot(kind="scatter" ,x = "Months_on_book" , y = "Avg_Utilization_Ratio")
```

```python
data = df["Attrition_Flag"].value_counts()
data.plot(kind="barh")
```

```python
df.groupby(["Education_Level", "Marital_Status"]).size().unstack().plot(kind="bar", stacked= True)
```
